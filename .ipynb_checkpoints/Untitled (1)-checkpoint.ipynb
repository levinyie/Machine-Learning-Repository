```python
import pandas as pd
```

```python
pd.options.display.max_columns=None
```

```python
df=pd.read_csv("credit_score_updated.csv")
```

```python
df.head()
```

```python
columns_to_check=["INCOME","SAVINGS","DEBT","CLOTHING","EDUCATION","ENTERTAINMENT","FINES","GAMBLING","GROCERIES","HEALTH","HOUSING","TAX","TRAVEL","UTILITIES","EXPENDITURE"]
```

```python
df.columns.tolist()
```

```python
df.columns=df.columns.str.strip()
```

```python
df["GAMBLING"].dtype
```

```python
for column in columns_to_check:
    df[column] = df[column].replace({'\$':''}, regex = True)
```

```python
df.head()
```

```python
for column in columns_to_check:
    df[column] = df[column].replace({'-':'0'}, regex = True)
```

```python
df.head(100)
```

```python
for column in columns_to_check:
    df[column] = df[column].replace({'\.':''}, regex = True)
    df[column] = df[column].replace({',':''}, regex = True)
```

```python
df.head()
```

```python
for column in columns_to_check:
    df[column] = pd.to_numeric(df[column])
    df[column] = df[column]/100
```
